# 1. 경사하강법 (Gradient Descent)

- y = wX + b

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(100)
y = X * 0.7

plt.figure(figsize=(8, 6))
plt.scatter(X, y)
```

```python
# 가중치
w = np.random.uniform(-1, 1)    # uniform : 균등분포
w
```

```python
# bias 값
b = np.random.uniform(-1, 1)
b
```

```python
lr = 0.5

# 딥러닝 아님 (신경망 없음)
# 모델 만들고 -> 손실 구하고 -> 최적화값 찾아서 역전파 -> 반복

for epoch in range(300) :
    y_pred = w*X + b                        # w*X + b : 모델

    # 1. 손실 구하기 (절대값(abs) or 제곱)
    error = np.abs(y_pred - y).mean()       # 손실함수 -> pyTorch 가면 함수로 있음(계산 안해도됨)
    
    # 2. 미분
    w_grad = lr * ((y_pred - y) * X).mean()
    b_grad = lr * (y_pred - y).mean()

    w = w - w_grad
    b = b - b_grad

    print(f"error : {error}")
    
    if epoch % 10 == 0 :            # 10번에 1번 확인
        y_pred = w*X + b
        plt.figure(figsize=(8, 6))
        plt.scatter(X, y)           # 실제정답
        plt.scatter(X, y_pred)      # 예측값
        plt.show()
    
    if error < 0.001 :              # early-stopping : 학습이 다 되기 전에 손실이 일정치에 도달하면 멈춤
        break
```
